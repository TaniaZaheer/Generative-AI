```python
from langchain import PromptTemplate
from langchain.llms import OpenAI
from langchain.chains import LLMChain
```

```python
study_plan_template = '''
Student Information:
- Name: {name}
- Field of Study: {field_of_study}
- Year of Study: {year_of_study}
- List of Subjects: {subjects}
- Academic Performance: {academic_performance}

Student Characteristics:
- Preferred Learning Styles: {learning_styles}
- Personal Objectives: {objectives}
- Challenges: {challenges}
- Extracurricular Activities: {extracurricular_activities}

Instructions:
1. Analyze the student's academic performance across different subjects.
2. Consider the student's preferred learning styles (visual, auditory, kinesthetic) when designing the study plan.
3. Take into account any personal objectives or challenges the student may have, such as preparing for specific exams or overcoming learning difficulties.
4. Must Incorporate extracurricular activities into the study timetable ensuring a balance between academic and non-academic commitments.
5. Ensure that the study plan is well-structured and comprehensive covering all subjects and providing sufficient study time for each.
6. Provide flexibility in the study plan to accommodate changes in the student's circumstances or feedback.
7. Generate the study plan based on the provided information and ensure it aligns with the student's unique needs.
'''
```

```python
study_plan_prompt = PromptTemplate(
    input_variables=["name", "field_of_study", "year_of_study", "subjects", "academic_performance", 
                     "learning_styles", "objectives", "challenges", "extracurricular_activities"],
    template=study_plan_template
)
llm = OpenAI(openai_api_key=" ", temperature=0.7)
study_plan_chain = LLMChain(llm=llm, prompt=study_plan_prompt)
```

```python
result = study_plan_chain.run({
    "name": "Tania Zaheer",
    "field_of_study": "Computer Science",
    "year_of_study": "4th Year",
    "subjects": "OOP, DSA, Web Development, Mobile Application Development",
    "academic_performance": "Good in Programming, Average in Algorithms, Excellent in Database Management",
    "learning_styles": "Visual, Kinesthetic",
    "objectives": "Improve coding skills, Prepare for internship interviews, Improve overall CGPA",
    "challenges": "Difficulty understanding complex algorithms,",
    "extracurricular_activities": "Playing Badminton every day atleast 1 hour"
})

print(result)
```
